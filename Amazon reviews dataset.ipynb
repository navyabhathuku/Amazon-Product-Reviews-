# Text Preprocessing

Remove Punctuations
Remove stopwords
Remove Lemmatization

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
```

```python

stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
corpus = []
for i in range(len(df)):
    rp=re.sub('[^a-zA-Z]',' ',df['reviews.text'][i])
    rp = rp.lower()
    rp = rp.split()
    rp = [wnl.lemmatize(word) for word in rp if not word in set(stopwords.words('english'))]

    rp = ' '.join(rp)
    corpus.append(rp)
    
print(corpus)    
   
```

# Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()

X = cv.fit_transform(corpus)
```

# Modelling using LDA

```python
from sklearn.decomposition import LatentDirichletAllocation
model = LatentDirichletAllocation()
model.fit(X)
```

```python
topic_results = model.transform(X)
```

```python
topic_results[0]
```

```python
topic_results[0].argmax()
```

# Combining with Original Data

```python
df['group'] = [topic_results.argmax(axis=1)for topic in topic_results]
```

```python
df.head()
```

# showing Top words per Topic

```python
for index, topic in enumerate(model.components_):
    print(f'The Top 10 WORDS For Topic #{index}')
    feature_names = list(cv.vocabulary_.keys())
    top_words_indices = topic.argsort()[-10:]
    top_words = [feature_names[i] for i in top_words_indices]
    print(top_words)
    print('\n')
```
